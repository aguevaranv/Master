## Metodología:

<div style="text-align: justify;">
Para el análisis se utiliza el proceso KDD que encaja con el propósito de este trabajo. El proceso KDD (Knowledge Discovery in Databases) es un proceso ampliamente utilizado en el análisis de datos y la extracción de conocimiento a partir de grandes conjuntos de datos. Fue propuesto por primera vez en la década de 1990 por los investigadores Fayyad, Piatetsky-Shapiro y Smyth.

El proceso de KDD consta de varias etapas, que se ejecutan de manera iterativa y cíclica para descubrir conocimiento útil y significativo a partir de los datos. Las etapas principales del proceso KDD son:

1. Selección de datos: En esta etapa, se identifican y seleccionan los conjuntos de datos relevantes para el análisis.

2. Preprocesamiento: En esta etapa, se realiza la limpieza de datos para eliminar valores atípicos, datos faltantes o duplicados. También puede implicar la normalización de los datos para asegurar que estén en un rango o formato adecuado.

3. Transformación: En esta etapa, los datos se convierten en una forma más adecuada para el análisis. Esto puede incluir la reducción de la dimensionalidad, la agregación de datos, la discretización u otras técnicas para representar los datos de manera efectiva.

4. Minería de datos: Es la etapa central del proceso KDD, donde se aplican algoritmos de minería de datos para descubrir patrones, relaciones, tendencias o modelos ocultos en los datos.

5. Evaluación: En esta etapa, se evalúan los resultados obtenidos de la minería de datos y se analizan e interpretan los patrones descubiertos.

6. Presentación del conocimiento: Una vez que se ha extraído el conocimiento, se presenta de manera comprensible y útil para los usuarios finales.
</div>

```python
# Extraer el mes y el año en columnas separadas
# loaded['mes'] = loaded['Mes y ano de ocurrencia'].str[:3]
loaded['ano'] = loaded['Mes y ano de ocurrencia'].str[-2:]
loaded['ano'] = pd.to_datetime(loaded['ano'], format="%y").dt.year
loaded = loaded.drop("Mes y ano de ocurrencia", axis=1)
display(loaded)
```

```python
# Calcular el total de fallecidos por año y por provincia 
total_provincias_anos = loaded.groupby('ano').agg('sum')
display(total_provincias_anos.T)
```

```python
# Identificar cuales son las 3 las provincias que más muertes ha tenido en los 9 años
total_provincias = total_provincias_anos.sum().sort_values(ascending=False).to_frame()
tres_provincias = total_provincias[:3].reset_index()
tres_provincias.columns = ["Provincia","Total"]
display(tres_provincias)
```

Representación gráfica de las provincias con el mayor numero de femicidios

```python
# - Representar graficamente las ciudades con más homicidios
import seaborn as sns
ax = sns.barplot(x='Provincia', y='Total', data=tres_provincias)
ax.set_title('Provincias con el mayor numero de homicidios')
```

```python
# Representar graficamente las 3 provincias con mas muertes, que muestre la evolucion en el tiempo por año
total_provincias_anos_reset_ndex = total_provincias_anos.T.reset_index()
total_provincias_anos_reset_ndex.columns = ['Provincia'] + total_provincias_anos_reset_ndex.columns[1:].tolist()
combined_dataset = tres_provincias.merge(right=total_provincias_anos_reset_ndex, on='Provincia', how='inner') # parametro 'on' define la columna pivote
combined_dataset = combined_dataset.drop("Total", axis=1)
display(combined_dataset)

plt.figure(figsize=(10, 6))

# Recorrer cada provincia y crear una línea para su evolución de años
for provincia in combined_dataset['Provincia']:
    data = combined_dataset[combined_dataset['Provincia'] == provincia].drop('Provincia', axis=1).values.flatten()
    plt.plot(combined_dataset.columns[1:], data, marker='o', label=provincia)

# Configurar los títulos y etiquetas de la gráfica
plt.title('Evolución del número de homicidios por año y provincia')
plt.xlabel('Año')
plt.ylabel('Número de homicidios')

# Mostrar la leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```

```python
#  - Traer los datos de femicidios y calcular el total de fallecidos por año y provincia tiene esos datos
ruta = os.path.join("data","femicidios.csv")
f_loaded = pd.read_csv(ruta, sep=';')
f_loaded.fillna(0).isna().sum()
# Extraer el mes y el año en columnas separadas
#f_loaded['mes'] = f_loaded['Mes y ano de ocurrencia'].str[:3]
f_loaded['ano'] = f_loaded['Mes y ano de ocurrencia'].str[-2:]
f_loaded['ano'] = pd.to_datetime(f_loaded['ano'], format="%y").dt.year
f_loaded = f_loaded.drop("Mes y ano de ocurrencia", axis=1)
display(f_loaded)
# Calcular el total de homicidios por año y por provincia 
f_total_provincias_anos = f_loaded.groupby('ano').agg('sum')
display(f_total_provincias_anos.T)
```

```python
# Identificar cuales son las 3 las provincias que más femicidios ha tenido en los 9 años

f_total_provincias = f_total_provincias_anos.sum().sort_values(ascending=False).to_frame()
f_tres_provincias = f_total_provincias[:3].reset_index()
f_tres_provincias.columns = ['Provincia','Total']
display(f_tres_provincias)
```

Representación gráfica de las provincias con el mayor numero de femicidios

```python
#  - Representar greficamente las provincias con más femicidios
import matplotlib.pyplot as plt
ax = sns.barplot(x='Provincia', y='Total', data=f_tres_provincias)
ax.set_title('Provincias con el mayor numero de femicidios')
plt.show()
```

Al comparar la gráfica de homicidios y femicidios se puede ver que las provincias de Guayas, Pichincha y Manabi son las provincias con mayor número de muertes de las 24 provicias del Ecuador.

Calcular el porcentaje de femicidios en cada año por cada provincia del total de homicidios

```python
# Calcular el porcentaje de femicidios en cada año por cada provincia del total de homicidios

total_fallecidos = f_total_provincias_anos * 100 / total_provincias_anos
#display(f_total_provincias_anos.T)
#display(total_provincias_anos.T)
total_fallecidos = total_fallecidos.fillna(0)
display(total_fallecidos.T)
```

```python
#  Representar graficamente el total de homicidios y su evolucion en el tiempo
fallecidos_ano = total_provincias_anos.sum(axis=1).to_frame().reset_index()
fallecidos_ano.columns = ['ano', 'homicidios']
display(fallecidos_ano)
sns.relplot(x='ano', y='homicidios', data=fallecidos_ano, kind='line')
# Obtener el objeto de la figura
fig = plt.gcf()

# Establecer el título de la gráfica
fig.suptitle('Evolución de los homicidios por año')
plt.show()
```

Se puede visualizar que el numero de homicidios aumenta drasticamente en el año 2021

Representar graficamente las evolución en el tiempo de los homicidios vs el femicidio

```python
#  Representar graficamente las evolución en el tiempo de los homicidios vs el femicidio

f_fallecidas_ano = f_total_provincias_anos.sum(axis=1).to_frame().reset_index()
f_fallecidas_ano.columns = ['ano', 'femicidios']

f_combined_dataset = fallecidos_ano.merge(right=f_fallecidas_ano, on='ano', how='inner') # parametro 'on' define la columna pivote
display(f_combined_dataset)

# Configurar el tamaño de la gráfica
plt.figure(figsize=(10, 6))

# Crear las líneas para los fallecidos y fallecidas
plt.plot(f_combined_dataset['ano'], f_combined_dataset['homicidios'], marker='o', label='homicidios')
plt.plot(f_combined_dataset['ano'], f_combined_dataset['femicidios'], marker='o', label='femicidios')

# Configurar los títulos y etiquetas de la gráfica
plt.title('Evolución del número de homicidios y femicidios por año')
plt.xlabel('Año')
plt.ylabel('Número de homicidios')

# Mostrar la leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```
