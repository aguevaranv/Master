```python
# Your solution HERE !!!!
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
ruta = os.path.join("data","homicidios.csv")
loaded = pd.read_csv(ruta, sep=';')
display(loaded)
```

```python
# Buscamos los campos que tienen NaN para cambiarlos por 0
loaded.isna().sum()
#display(loaded[loaded.isna()])
```

```python
# Reemplazo nan por 0
loaded.fillna(0).isna().sum()
```

```python
# Extraer el mes y el año en columnas separadas
# loaded['mes'] = loaded['Mes y ano de ocurrencia'].str[:3]
loaded['ano'] = loaded['Mes y ano de ocurrencia'].str[-2:]
loaded['ano'] = pd.to_datetime(loaded['ano'], format="%y").dt.year
loaded = loaded.drop("Mes y ano de ocurrencia", axis=1)
display(loaded)
```

```python
# Calcular el total de fallecidos por año y por provincia 
total_provincias_anos = loaded.groupby('ano').agg('sum')
display(total_provincias_anos.T)
```

```python
# Identificar cuales son las 3 las provincias que más muertes ha tenido en los 9 años
total_provincias = total_provincias_anos.sum().sort_values(ascending=False).to_frame()
tres_provincias = total_provincias[:3].reset_index()
tres_provincias.columns = ["Provincia","Total"]
display(tres_provincias)
```

Representación gráfica de las provincias con el mayor numero de femicidios

```python
# - Representar graficamente las ciudades con más homicidios
import seaborn as sns
ax = sns.barplot(x='Provincia', y='Total', data=tres_provincias)
ax.set_title('Provincias con el mayor numero de homicidios')
```

```python
#  - Traer los datos de femicidios y calcular el total de fallecidos por año y provincia tiene esos datos
ruta = os.path.join("data","femicidios.csv")
f_loaded = pd.read_csv(ruta, sep=';')
f_loaded.fillna(0).isna().sum()
# Extraer el mes y el año en columnas separadas
#f_loaded['mes'] = f_loaded['Mes y ano de ocurrencia'].str[:3]
f_loaded['ano'] = f_loaded['Mes y ano de ocurrencia'].str[-2:]
f_loaded['ano'] = pd.to_datetime(f_loaded['ano'], format="%y").dt.year
f_loaded = f_loaded.drop("Mes y ano de ocurrencia", axis=1)
display(f_loaded)
# Calcular el total de fallecidos por año y por provincia 
f_total_provincias_anos = f_loaded.groupby('ano').agg('sum')
display(f_total_provincias_anos.T)
```

```python
# Identificar cuales son las 3 las provincias que más femicidios ha tenido en los 9 años

f_total_provincias = f_total_provincias_anos.sum().sort_values(ascending=False).to_frame()
f_tres_provincias = f_total_provincias[:3].reset_index()
f_tres_provincias.columns = ['Provincia','Total']
display(f_tres_provincias)
```

Representación gráfica de las provincias con el mayor numero de femicidios

```python
#  - Representar greficamente las ciudades con más femicidios
import matplotlib.pyplot as plt
ax = sns.barplot(x='Provincia', y='Total', data=f_tres_provincias)
ax.set_title('Provincias con el mayor numero de femicidios')
plt.show()
```
